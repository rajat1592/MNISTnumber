```python
import numpy as np
import matplotlib.pyplot as plt
import tensorflow
from tensorflow import keras #Keeras handales layers,activation function and optimizers
from tensorflow.keras import Sequential  #This is used for creating Sequential Nuralnetwork
from tensorflow.keras.layers import Dense,Flatten
```

```python
(X_train,y_train),(X_test,y_test)=keras.datasets.mnist.load_data()
```

```python
plt.imshow(X_train[0])
print('this is the image of '+ str(y_train[0]))
```

```python
X_train=X_train/255
X_test=X_test/255
```

```python
model=Sequential()   
model.add(Flatten(input_shape=(28,28)))   #the data should be in 1-d form
model.add(Dense(300,activation='relu'))   #here i use 300 perceptrons whic have Relu as activation function
                                          #[we can also use sigmoid,tanh,leaky relu,softmax]
model.add(Dense(10,activation='softmax')) #Generaly for multiclass classification we use Softmax in outter layer
```

```python
model.summary()
```

```python
model.compile(loss='sparse_categorical_crossentropy',optimizer='Adam')  #we can experiment with different optimization techiniques
                                                                        #[Gradiant descent,Stochiastic Gradiant Descent,adam,momentum]
```

```python
model.fit(X_train,y_train,epochs=15,validation_split=0.2)
```

```python
y_prob = model.predict(X_test)
```

```python
y_pred=y_prob.argmax(axis=1)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
```
